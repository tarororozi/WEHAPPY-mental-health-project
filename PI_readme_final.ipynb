```python
import os
import pymongo

import os
import pymongo

# Establish connection to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
# Access or create database
db = client["user_database"]
# Access or create collection within the database
collection = db["users"]

# Function to read files in a folder and insert their content into MongoDB
def read_files(folder_path):
    # Iterate through all files in the folder and its subfolders
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # Check if the file ends with '_readme.txt'
            if file.endswith('_readme.txt'):
                # Get the full path of the file
                file_path = os.path.join(root, file)
                # Open the file and read its content
                with open(file_path, 'r') as f:
                    content = f.read()
                    # Insert file path and content into MongoDB
                    collection.insert_one({"file_path": file_path, "content": content})
                # Optionally print the inserted file path
                #print(f"Inserted file: {file_path}")

# Function to generate folder paths for a range of subjects
def generate_folder_paths(base_folder, start_index, end_index):
    folder_paths = []
    # Iterate through the range of subjects
    for i in range(start_index, end_index + 1):
        # Create folder path for each subject and append to list
        folder_paths.append(os.path.join(base_folder, f'S{i}'))
    return folder_paths

# Generate folder paths for subjects 2 to 11
folder_paths = generate_folder_paths('/Users/talia/Desktop/infos/columbia/5400py/untitled folder/WESAD/', 2, 11)
print(folder_paths)

# Iterate through each folder path and read files within them
for folder_path in folder_paths:
    read_files(folder_path)

# Print message indicating successful insertion of data into MongoDB
print("Data inserted successfully into MongoDB.")

```

```python
data_list = []

# Iterate through each document in the collection
for document in collection.find():
    document_data = {}

    # Split the content of the document by lines
    lines = document['content'].split('\n')

    # Iterate through each line of the content
    for line in lines:
        # Check if the line contains a colon (:)
        if ':' in line:
            # Split the line into key-value pair
            key, value = line.split(':', 1)
            # Remove leading and trailing spaces and store in document_data dictionary
            document_data[key.strip()] = value.strip()
        # Check if the line contains a question mark (?)
        elif '?' in line:
            # Split the line into key-value pair
            key, value = line.split('?', 1)
            # Remove leading and trailing spaces and store in document_data dictionary
            document_data[key.strip()] = value.strip()

    # Append the document_data dictionary to the data_list
    data_list.append(document_data)

# Convert the data_list into a DataFrame
df = pd.DataFrame(data_list)

# Rename columns as per the specified mapping
df.rename(columns={
    'Age': 'age',
    'Height (cm)': 'height',
    'Weight (kg)': 'weight',
    'Gender': 'gender',
    'Dominant hand': 'dominant_hand',
    'Did you drink coffee today': 'coffee_today',
    'Did you drink coffee within the last hour': 'coffee_last_hour',
    'Did you do any sports today': 'sport_today',
    'Are you a smoker': 'smoker',
    'Did you smoke within the last hour': 'smoke_last_hour',
    'Do you feel ill today': 'feel_ill_today'
}, inplace=True)

# Reorder columns as per the specified sequence
df = df[['age', 'height', 'weight', 'gender', 'dominant_hand',
         'coffee_today', 'coffee_last_hour', 'sport_today',
         'smoker', 'smoke_last_hour', 'feel_ill_today']]

# Print the DataFrame
print(df)
```

```python
import psycopg
#create connections
conn =  psycopg.connect(
host="localhost",
port='5432',
dbname="5400api",
user="postgres",
password="123")

cur = conn.cursor()  
def create_table_with_patient_id():
       

    # from the number 2 since first patient id is S2
    cur.execute("CREATE SEQUENCE user_data_seq_id START 2;")

    # Create the user_data table with the new patient_id column
    cur.execute("""
    CREATE TABLE IF NOT EXISTS user_data (
        patient_id VARCHAR(50) PRIMARY KEY,
        age INT,
        height REAL,
        weight REAL,
        gender VARCHAR(6),
        dominant_hand VARCHAR(5),
        coffee_today VARCHAR(3),
        coffee_last_hour VARCHAR(3),
        sport_today VARCHAR(3),
        smoker VARCHAR(3),
        smoke_last_hour VARCHAR(3),
        feel_ill_today VARCHAR(3)
    );
    """)

    # Create a function to generate patient_id
    cur.execute("""
    CREATE OR REPLACE FUNCTION set_patient_id()
    RETURNS TRIGGER AS $$
    BEGIN
      NEW.patient_id := 'S' || nextval('user_data_seq_id');
      RETURN NEW;
    END;
    $$ LANGUAGE plpgsql;
    """)

    # Create a trigger that will call the set_patient_id function before an insert operation
    cur.execute("""
    CREATE TRIGGER trigger_set_patient_id
    BEFORE INSERT ON user_data
    FOR EACH ROW
    EXECUTE FUNCTION set_patient_id();
    """)

    # Commit changes
    conn.commit()


# Function call to create the table with patient_id
create_table_with_patient_id()
```

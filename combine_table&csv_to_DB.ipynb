```python
csv_to_table = {}
for i in range(len(table_name)):
    csv_to_table[file_dir[i]] = table_name[i]

print(csv_to_table)
```

Rename the column name in the csv, since there all have same column name originally.

```python
import pandas as pd
import os

# Directory containing CSV files
directory = 'pickle_raw'

# Get list of files in the directory
files = os.listdir(directory)

# Initialize a dictionary to store DataFrames by prefix
dataframes = {}

table_names = [
    'EDA_wrist',
    'BVP_wrist',
    'ACC_wrist',
    'TEMP_wrist',
    'label',
    'Resp',
    '_EDA_wrist',
    '_BVP_wrist',
    '_ACC_wrist',
    '_TEMP_wrist',
    '_label',
    '_Resp'
]

# Iterate through files and group DataFrames by prefix
for file in files:
    if file.endswith('.csv'):
        table_name = file.split('.')[0][3:]
        #print(table_name)
        prefix = file.split('_')[0]
        print(prefix)
        if table_name in table_names:
            print(1)
            if prefix not in dataframes:
                dataframes[prefix] = []
            file_path = os.path.join(directory, file)
            #print(file_path)
            df = pd.read_csv(file_path)
            df.columns = [f'{file.split(".")[0]}_{col}' for col in df.columns]
            print(df.columns)
            dataframes[prefix].append(df)
```

Create a dictonary store all the dataframe

```python
# Concatenate DataFrames with the same prefix
combined_dfs = {}
for prefix, dfs in dataframes.items():
    combined_df = pd.concat(dfs, axis=1)
    combined_dfs[prefix] = combined_df
```

```python
combined_dfs
```

Combine the dataframe for the same participants and save it

```python
for prefix, df in combined_dfs.items():
    df = pd.DataFrame(df)
    df.to_csv(f'{prefix}_combined_data.csv', index=False)
```

## Start to store the csv into the database

```python
import os
import pandas as pd
import psycopg2

# PostgreSQL database connection parameters
db_params = {
    'host': 'localhost',
    'port': '5432',
    'database': '5400-final',
    'user': 'postgres',
    'password': '123'
}

# Directory containing CSV files
directory = 'pickle_raw'

# Establish database connection
conn = psycopg2.connect(**db_params)
cursor = conn.cursor()

# Get list of files in the directory
files = os.listdir(directory)

# Load CSV files into DataFrames and create tables in PostgreSQL
for file in files:
    if file.endswith('.csv'):
        # Extract table name from file name (assuming file name matches table name)
        table_name = file.split('.')[0]
        
        # Read CSV file into DataFrame
        file_path = os.path.join(directory, file)
        df = pd.read_csv(file_path)
        
        # Create table in PostgreSQL
        create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} ({', '.join([f'{col} TEXT' for col in df.columns])})"
        cursor.execute(create_table_query)
        
        # Insert data into PostgreSQL table
        df_columns = ', '.join(df.columns)
        for row in df.itertuples(index=False):
            # Replace NaN values with NULL in the INSERT statement
            values = ', '.join([f"'{value}'" if pd.notna(value) and isinstance(value, str) else 'NULL' if pd.isna(value) else str(value) for value in row])
            insert_query = f"INSERT INTO {table_name} ({df_columns}) VALUES ({values})"
            cursor.execute(insert_query)

# Commit changes and close connection
conn.commit()
cursor.close()
conn.close()

```
